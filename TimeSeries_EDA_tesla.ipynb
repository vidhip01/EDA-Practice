### Time Resampling

```python
df_tesla.tail()
```

```python
df_tesla= df_tesla.set_index('Date', drop=True)
```

```python
df_tesla.head()
```

Rule A means year and frequency

```python
df_tesla.resample(rule="A")
```

```python
df_tesla.resample(rule="A").min()
```

it is giving minum price of all columns 

```python
df_tesla.resample(rule="A").max()
```

now maximum price

```python
type(df_tesla.resample(rule="A").max())
```

```python
df_tesla.resample(rule="A").max()['Open'].plot()
```

Rule QS if for quaterly start frequency


```python
df_tesla.resample(rule="QS").max()
```

```python
df_tesla.resample(rule="QS").max()['Open'].plot()
```

```python
df_tesla.resample(rule="QS").max()['High'].plot()
```

Rule BS means Business End Frequency

```python
df_tesla.resample(rule="BA").max()
```

```python
df_tesla.resample(rule="BA").max()['High'].plot()
```

Rule BQS means Business Quarters

```python
df_tesla.resample(rule="BQS").max()
```

```python
df_tesla.resample(rule="QS").max()['Open'].plot()
```

### Plotting resampling data

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

Rule M means every month data

```python
df_tesla['Open'].resample(rule='M').mean()
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar', figsize=(15,6))
```

```python
figure.autofmt_xdate()
df_tesla['Open'].resample(rule='M').max().plot(kind='bar', figsize=(20,6))
# Add labels and title
plt.xlabel('Date')
plt.ylabel('Maximum Opening Price')
plt.title('Maximum Monthly Opening Price of Tesla Stock')
```
